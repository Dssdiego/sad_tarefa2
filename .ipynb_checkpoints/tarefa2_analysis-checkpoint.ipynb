```python
# Imports

# Current Google Trends: https://trends.google.com.br/trends/explore?date=2020-01-01%202020-12-31&geo=BR&q=%2Fg%2F11h10_t1nf,%2Fg%2F11flpjdym5,%2Fm%2F0130pygf

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from pytrends.request import TrendReq
%matplotlib inline
```

```python
# Timeframes
timeframe1Ano = '2020-01-01 2020-12-31'
timeframe5Anos = '2016-01-01 2020-12-31'
timeframefinalano = '2020-11-01 2020-12-31'
```

```python
# Interesse 1 ano
trend = TrendReq()
trend.build_payload(
    kw_list=['PlayStation 5', 'Xbox Series X', 'Nintendo Switch'],
    timeframe=timeframe1Ano,
    geo='BR'
)

df = trend.interest_over_time()
df = df.drop(axis=1,labels='isPartial') # Remove a coluna 'isPartial'
df.plot(kind='line', figsize=(10,5)).set(
    title='Interesse ao longo do tempo (01/01/2020 - 31/12/2020)',
    ylabel='Interesse (em pontos)',
    xlabel='Data'
)
# plt.savefig('img/interesse_1ano.png')
```

```python
# Interesse 5 anos
trend = TrendReq()
trend.build_payload(
    kw_list=['PlayStation 5', 'Xbox Series X', 'Nintendo Switch'],
    timeframe=timeframe5Anos,
    geo='BR'
)

df = trend.interest_over_time()
df = df.drop(axis=1,labels='isPartial') # Remove a coluna 'isPartial'
df.plot(kind='line', figsize=(10,5)).set(
    title='Interesse ao longo do tempo (01/01/2016 - 31/12/2020)',
    ylabel='Interesse (em pontos)',
    xlabel='Data'
)
# plt.savefig('img/interesse_5anos.png')
```

```python
# Interesse Final Ano
trend = TrendReq()
trend.build_payload(
    kw_list=['PlayStation 5', 'Xbox Series X', 'Nintendo Switch'],
    timeframe=timeframefinalano,
    geo='BR'
)

df = trend.interest_over_time()
df = df.drop(axis=1,labels='isPartial') # Remove a coluna 'isPartial'
df.plot(kind='line', figsize=(10,5)).set(
    title='Interesse ao longo do tempo (01/11/2020 - 31/12/2020)',
    ylabel='Interesse (em pontos)',
    xlabel='Data'
)

x = range(len(df.index))

# y = df['Xbox Series X']
# f = np.poly1d(np.polyfit(x,y,1))

# plt.plot(df.index, f(x), 'r', label='Xbox Series X')

# y = df['Nintendo Switch']
# f = np.poly1d(np.polyfit(x,y,1))

# plt.plot(df.index, f(x), 'orange', label='Nintendo Switch')

y = df['PlayStation 5']
f = np.poly1d(np.polyfit(x,y,1))

plt.plot(df.index, f(x), 'k.')

plt.legend()
plt.savefig('img/final_ano.png')
```

```python
trend = TrendReq()
trend.build_payload(
    kw_list=['PlayStation 5', 'Xbox Series X', 'Nintendo Switch'],
    timeframe=timeframe1Ano,
    geo='BR'
)

df = trend.interest_over_time()
df = df.drop(axis=1,labels='isPartial') # Remove a coluna 'isPartial'
# df.reset_index(level=0, inplace=True)
# df['date'] = pd.to_datetime(df['date']).astype('datetime64[ns]')
# df.loc[to_datetime(datetime.fromisoformat(row['timestamp']))] = data
```

```python
y = df['Xbox Series X']
x = range(len(df.index))

f = np.poly1d(np.polyfit(x,y,1))

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8,6), sharex=True, sharey=True)
fig.tight_layout()

ax1.plot(df.index, y, 'o', label='Dados', color='#1f77b4')
ax1.plot(df.index, f(x), 'r', label='Tendência')
ax1.set(title='Xbox Series X')
ax1.legend()

y = df['PlayStation 5']
f = np.poly1d(np.polyfit(x,y,1))

ax2.plot(df.index, y, 's', label='Dados', color='#9467bd')
ax2.plot(df.index, f(x), 'r', label='Tendência')
ax2.set(title='PlayStation 5')
ax2.legend()

y = df['Nintendo Switch']
f = np.poly1d(np.polyfit(x,y,1))

ax3.plot(df.index, y, 'D', label='Dados', color='#17becf')
ax3.plot(df.index, f(x), 'r', label='Tendência')
ax3.set(title='Nintendo Switch')
ax3.legend()

# plt.savefig('img/tendency.png')
```

```python
# Investidor Conservador
my_data = [80,10,10]
my_labels = 'Nintendo Switch','Xbox Series X', 'PlayStation 5'
plt.pie(my_data,labels=my_labels,autopct='%1.1f%%')
plt.title('Distribuição dos Investimentos')
plt.axis('equal')
plt.xlabel('Perfil: Conservador')
plt.savefig('img/inv_1.png')
plt.show()
```
