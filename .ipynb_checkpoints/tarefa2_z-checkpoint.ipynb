```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from pytrends.request import TrendReq
%matplotlib inline

# Timeframe: 12 meses
timeframe1Ano = '2020-01-01 2020-12-31'

# Declaração dos tipos de dados
trend = TrendReq()
trend.build_payload(
    kw_list=['PlayStation 5', 'Xbox Series X', 'Nintendo Switch'],
    timeframe=timeframe1Ano,
    geo='BR'
)
```

```python
df = trend.interest_over_time()
df = df.drop(axis=1,labels='isPartial') # Remove a coluna 'isPartial'
df.head()
```
